```python
#
# Attempt at an unfair wheel at casino
#
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
#number of trials for the simulation
nran = 100000 
#
#array to store the results of the roulette spins
# Using integers as data type for r_draw array
r_draw_UF = np.zeros(nran, dtype='i')
#
#Seed the random number generator for reproducibility
random.seed(a=121, version=2)
#
# Simulate the unfair roulette spins
# Using random.uniform to generate random spins between 1 and 37
for x in range(nran):
    r_draw_UF[x] = random.uniform(1, 38)
#
#Initialize an array to track winnings or losses for each spin
win_lose_UF = np.zeros(nran, dtype='i')
#
# number you are betting on
bet = 8
#
for x in range(nran):
    if bet == r_draw_UF[x]:
        win_lose_UF[x] = 35  # Win 35 times your bet
    else:
        win_lose_UF[x] = -1  # Lose your bet
#
# Calculate the total win/loss
win_total = sum(win_lose_UF)
print('Unfair Roulette Results:')
win_pl = round(win_total / nran, 3)
print(f"Total Profit/Loss per spin: {win_pl}")
#
#Calculate the average and standard deviation
win_average = round(np.mean(win_lose_UF), 2)
win_sd = round(np.std(win_lose_UF), 4)
print(f"Average Profit/Loss: {win_average}")
print(f"Standard Deviation: {win_sd}")
#
# Calculate confidence interval bands
# Using a 95% confidence interval (1.96 corresponds to 95% for a normal distribution)
bands = round(1.96 * win_sd / (nran**0.5), 4)
print(f"Confidence Interval Band: {bands}")
#
# confidence interval
ci_low = round(win_average - bands, 4)
ci_high = round(win_average + bands, 4)
print(f"Confidence Interval: [{ci_low}, {ci_high}]")
```

```python
#
# histogram for the outcomes of the unfair roulette spins
plt.hist(r_draw_UF, bins=37, color='red', alpha=0.7)  # 37 bins for numbers 1 through 37
plt.title("Histogram of Unfair Roulette Spins")
plt.xlabel("Roulette Number")
plt.ylabel("Frequency")
plt.show()
#
# histogram for win/loss outcomes
plt.hist(win_lose_UF, bins=50, color='blue', alpha=0.7)  # 50 bins to see more detail on wins/losses
plt.title("Histogram of Winnings/Losses per Spin")
plt.xlabel("Profit/Loss")
plt.ylabel("Frequency")
plt.show()
```
